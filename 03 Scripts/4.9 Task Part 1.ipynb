## 5. Completing fundamental data quality and consistency checks.

```python
# Checking for mixed_type data.
for col in df_cx.columns.tolist():
  weird = (df_cx[[col]].applymap(type) != df_cx[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cx[weird]) > 0:
    print (col)
```
