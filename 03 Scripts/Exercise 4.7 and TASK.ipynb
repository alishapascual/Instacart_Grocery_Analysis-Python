## If-Statements with User-Defined Functions

```python
# Defining a path to Instacart Project folder.
path = r'/Users/terencekeys/Downloads/Shawzz Folder/06-2023 Instacart Basket Analysis'
```

```python
# Importing orders_products_merged dataframe
df_ords_prods_merged =  pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data','orders_products_merged.pkl'))
```

```python
# Creating a subset of the first 1 million rows for the ords_prods_merge dataframe.
df = df_ords_prods_merged[:1000000]
```

```python
# Checking the shape of the new subset.
df.shape
```

```python
# Defining price label function.
def price_label(row):

  if row['prices'] > 5 and (row['prices'] <= 15):
    return 'Mid-range product'
  elif (row['prices'] <= 5):
    return 'Low-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
# Checking the values.
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking the values.
df['price_range'].value_counts(dropna = False)
```

```python
# Checking the highest price
df['prices'].max()
```

## If-Statements with the loc() Function

```python
# Using loc() function on subset.
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Using loc() function on dataframe.
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

## If-Statements with For-Loops

```python
# Seeing the shape/dimension and the column names
df_ords_prods_merged
```

```python
# Getting value counts of every day of the week.
df_ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
# Finding busiest, regularly busy, and least busy day.
result = []

for value in df_ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy day")
  else:
    result.append("Regularly busy day")
```

```python
# Getting the result
result
```

```python
# Creating a new column for result in dataframe.
df_ords_prods_merged['busiest day'] = result
```

```python
# Showing value count for busiest, regularly busy, and least bust day.
df_ords_prods_merged['busiest day'].value_counts(dropna = False)
```
